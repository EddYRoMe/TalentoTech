```python
'''Algoritmo para comparar las áreas de un triángulo
y un cuadrado, al final debe indicar cual de ellas es mayor '''
def Comparador(Base_T, Altura_T, Lado_C):
  Base_T= int(input('Ingrese el valor de la base del triángulo: ')) # Pido ingreso por teclado de los parámetros del triángulo
  Altura_T= int(input('Ingrese el valor de la altura del triángulo: '))

  Area_T= float(Base_T*Altura_T/2) #Calculo el área del triángulo

  Lado_C= int(input('Ingrese el valor del lado del cuadrado: ')) # Pido ingreso por teclado de los parámetros del triángulo
  Area_C= float(Lado_C**2) #Calculo el área del área del cuadrado

  if Area_T > Area_C:
    print('El área del triángulo es mayor')
    print ('Las áreas son: {} el cuadrado y {} el triángulo'.format(Area_C,Area_T))
  else:
    print('El área del cuadrado es mayor')
    print ('Las áreas son: {} el cuadrado y {} el triángulo'.format(Area_C,Area_T))

```

```python
from scipy.stats import norm
dist = norm(loc=np.mean(df_hurtoPers['edad']), scale=np.std(df_hurtoPers['edad']))
x = np.linspace(dist.ppf(0.001),
                dist.ppf(0.999), 100)
plt.hist(df_hurtoPers['edad'], density=True, label="Muestra", color="rebeccapurple", edgecolor = "black") # Changed 'normed' to 'density'
plt.plot(x, dist.pdf(x), "k p", label="PDF teórico")
plt.legend()
```
